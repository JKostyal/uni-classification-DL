# University Detector
Detects which university is which - between Harvard, Stanford, and Georgia State. ~92% Accuracy.

```python
from fastai.vision.all import *
from fastai.vision.widgets import *
import voila
```

```python
path = Path()
learn_inf = load_learner(path/'uni_export.pkl',cpu = True)
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
```

```python
def on_new_upload(change):
    lbl_pred.value = ''
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Predction: {pred}, Probability: {probs[pred_idx]:.04f}'
```

```python
btn_upload.observe(on_new_upload,names = ['data'])
```

```python
display(VBox([widgets.Label('Select your university'), btn_upload, out_pl, lbl_pred]))
```
